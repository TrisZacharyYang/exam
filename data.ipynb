```python
import matplotlib.pyplot as plt
import pandas as pd

#1
df = pd.read_csv('rentals.csv')
df['price'] = df['price'].str.strip().str.replace('$','', regex = False).str.replace(',','', regex = False).astype(float)

print(df.groupby(['property_type'])['price'].mean()) 



#2
data = df.groupby(['property_type'])['price'].value_counts()
data = data.to_frame()

plt.figure(figsize =(40, 7))
plt.boxplot(data, labels = data.index)
plt.show()
```
